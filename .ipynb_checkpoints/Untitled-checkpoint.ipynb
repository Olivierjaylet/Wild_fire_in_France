```python
csv_file ="liste_incendies_ du_12_08_2022.csv"
dataframe = dataclean(csv_file)
```

```python
year=2022
```

```python
grouped_dt = dataframe[(dataframe['Année'] == year)].groupby((dataframe['Département'])).sum('Surface parcourue (ha)')
```

```python
grouped_dt
```

```python
total = grouped_dt['Surface parcourue (ha)'].sum()
```

```python
grouped_dt['pourcentage'] = (grouped_dt['Surface parcourue (ha)']/total)*100
```

```python
grouped_dt
```

```python
counts = grouped_dt['pourcentage']

# Créez un camembert à partir des comptages
plt.figure(figsize=(6, 6))  # Taille du camembert
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Répartition des Catégories')  # Titre du camembert
plt.axis('equal')  # Pour s'assurer que le camembert est un cercle

# Affichez le camembert
plt.show()
```
